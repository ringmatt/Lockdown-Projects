```python
# Packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.style as style
from datetime import datetime
import scipy.stats as stats
style.use('fivethirtyeight')
```

```python
# Checking the encoding of the data
with open("../../Data/PlaylistData/ScanlanShorthalt.csv") as f:
    print(f)
```

```python
# Importing the data
dfScanlan = pd.read_csv("../../Data/PlaylistData/ScanlanShorthalt.csv", encoding = "cp1252")
dfCatharsis = pd.read_csv("../../Data/PlaylistData/Catharsis.csv", encoding = "cp1252")
dfCarefree = pd.read_csv("../../Data/PlaylistData/Carefree.csv", encoding = "cp1252")
dfSelfLove = pd.read_csv("../../Data/PlaylistData/SelfLove.csv", encoding = "cp1252")
dfThankful = pd.read_csv("../../Data/PlaylistData/Thankful.csv", encoding = "cp1252")
dfLove = pd.read_csv("../../Data/PlaylistData/Love.csv", encoding = "cp1252")
dfNostalgia = pd.read_csv("../../Data/PlaylistData/Nostalgia.csv", encoding = "cp1252")
dfX = pd.read_csv("../../Data/PlaylistData/X.csv", encoding = "cp1252")
playlists = [ dfCatharsis, dfSelfLove, dfLove, dfNostalgia, dfX] # dfScanlan, dfThankful, dfCarefree, 
playlistName = ["Catharsis", "Self Love", "Love", "Nostalgia", "X"] # "Scanlan Shorthalt", "Thankful", "Carefree", 
```

```python
# Changing "Length" to a float in each dataset
i = 0
for playlist in playlists:
#     print(playlistName[i])
    playlist["Length"] = playlist.apply(lambda row: datetime.strptime(row["Length"], '%M:%S'), axis = 1)
    playlist["Length"] = (playlist["Length"].dt.minute * 60 + playlist["Length"].dt.second)
    i += 1
```

```python
# Merging into one Dataset
df = pd.DataFrame(columns = dfScanlan.columns)

i = 0
for playlist in playlists:
    playlist["Playlist"] = playlistName[i]
    df = df.append(playlist)
    i += 1

# Changes relevant attributes to floats
attributes = ["BPM", "Energy", "Dance", "Loud", "Valence", "Length", "Acoustic", "Popularity"]
for atr in attributes:
    df[atr] = df[atr].astype(float)
    
df.head()
```

```python
df.to_csv("../../Data/PlaylistData/CombinedPlaylists.csv", index = False)
```

```python
plt.figure(figsize = (15,5))

i = 0
for playlist in playlists:
    sns.kdeplot(playlist["BPM"], label = playlistName[i])
    i += 1

plt.legend(loc = "upper right")
```
