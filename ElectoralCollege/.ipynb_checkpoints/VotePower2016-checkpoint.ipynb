### Voting Power, 2016
The purpose of this report is to determine how many Trump votes translated to an electoral college vote, and the same for Clinton. A ratio will then be made to determine the relative power of Trump to Clinton votes. Questions to be answered are as follows:

1. How many votes for Trump translated to an Electoral College vote, and the same for Clinton?
2. How has the voting power of states changed over the years?
3. How has the electoral college shifted over the years?

```python
import pandas as pd
```

```python
# Importing the main dataset
df = pd.read_csv("1976-2016-president.csv")
df.head()
```

```python
# Importing the Electoral College dataset
ec = pd.read_excel("ElectoralVotes.xlsx")
ec.tail()
```

### Cleaning the Data

All "NaN"s (Missing data) will be removed. This appears to mostly be the "notes" column, write in candidates, and some 3rd parties.

```python
# Removing the notes column, which is all empty
df = df.drop(columns = ["notes"])
```

```python
# Removing all NaNs
df = df.dropna().reset_index(drop = True)
df.tail()
```

### Preparing the Data

Here I'm creating useful statistics for later analysis and exploration.

```python
import seaborn as sns
```

```python
# Creating a data table of the total number of EC votes per year
totalECVotes = ec.groupby("year")["Electoral Votes"].sum().to_frame().reset_index()
totalECVotes = totalECVotes.rename(index = str, columns = {"Electoral Votes": "Total EC Votes"})

# Merging with the original dataframe
ec = ec.merge(totalECVotes, on = 'year')
ec.tail()
```

```python
# Creating a column of the percentage of the total EC votes
ec["Percent of EC Votes"] = ec["Electoral Votes"]/ec["Total EC Votes"]
ec.tail()
```

```python
# Merge this with the main dataset
df = df.merge(ec, on = ["year", "state"])
```

```python
# Remove the totals columns
df = df.drop(labels = ["Total EC Votes"], axis = 1)
```

```python
df.tail()
```

```python
sns.lineplot(x = 'year', y = 'Percent of EC Votes', hue = 'state', data = df)
```

```python
import matplotlib.pyplot as plt
import matplotlib.style as style
style.use('fivethirtyeight')
```

```python
for state in df['state'][df['Percent of EC Votes'] >= 0.035].unique():
    plt.plot(df[df['state'] == state]["year"], df[df['state'] == state]["Percent of EC Votes"], label = state)
    if state == "Pennsylvania" or state == "Florida":
        plt.text(df[df['state'] == state]["year"].iloc[-1], df[df['state'] == state]["Percent of EC Votes"].iloc[-1]+0.005, state)
    else:
        plt.text(df[df['state'] == state]["year"].iloc[-1], df[df['state'] == state]["Percent of EC Votes"].iloc[-1], state)
plt.legend(bbox_to_anchor = (1.1,1.0), fontsize = 16)
plt.ylabel("Portion of Electoral College Votes")
plt.xlabel("Year")
plt.title("Change in Electoral College")
```

```python
# Dividing state population by the number of electoral votes per year
df["Popular Votes per Electoral Vote"] = df["totalvotes"]/df["Electoral Votes"]
```

```python
# Grouping by state and year to make a list
votePower = df.groupby(["year", "state", "Popular Votes per Electoral Vote"]).count().reset_index()[["year", "state", "Popular Votes per Electoral Vote"]]
votePower = votePower.rename(index = str, columns = {"Popular Votes per Electoral Vote":"VotesPerEV"})
```

```python
# Wyoming Standard
votePower["Number of Wyoming Votes"] = 0
for i in range(len(votePower)):
    votePower["Number of Wyoming Votes"].iloc[i] = votePower[votePower["year"] == votePower["year"].iloc[i]][votePower["state"] == "Wyoming"]["VotesPerEV"].iloc[0]/votePower["VotesPerEV"].iloc[i]
votePower.tail()
```

```python
for state in votePower['state'][votePower['Number of Wyoming Votes'] <= 0.28].unique():
    plt.plot(votePower[votePower['state'] == state]["year"], votePower[votePower['state'] == state]["Number of Wyoming Votes"], label = state)
#     if state == "Pennsylvania" or state == "Florida":
#         plt.text(df[df['state'] == state]["year"].iloc[-1], df[df['state'] == state]["Percent of EC Votes"].iloc[-1]+0.005, state)
#     else:
    plt.text(votePower[votePower['state'] == state]["year"].iloc[-1], votePower[votePower['state'] == state]["Number of Wyoming Votes"].iloc[-1], state)
plt.legend(bbox_to_anchor = (1.1,1.0), fontsize = 16)
plt.ylabel("Value of Vote Compared to WY")
plt.xlabel("Year")
plt.title("Wyoming Standard")
```
